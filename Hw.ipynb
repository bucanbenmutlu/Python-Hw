```python
weight = float(input("Enter your weight (in kilograms )"))
height = float(input("Enter your height (in meters ) "))

bmi = weight / (height ** 2)


if bmi <= 18.5:
  print("Under weight !")
elif 18.5 < bmi <= 24.9:
  print("Healthy weight !")
elif 25 <= bmi <= 29.9:
  print("Over weight !")
elif bmi >= 30:
  print("Obesity !")
```
